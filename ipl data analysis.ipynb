```python
balls = df.groupby('batsman')['ball'].count().reset_index()
balls
```

```python
runs = df.groupby('batsman')['batsman_runs'].sum().reset_index()
runs
```

```python
fours= df[df['batsman_runs']==4]
fours = fours.groupby('batsman')['batsman_runs'].count().reset_index()
fours = fours.rename(columns={'batsman':'batsman','batsman_runs':'4s'})
fours.head()
```

```python
sixes = df[df['batsman_runs']==6]
sixes = sixes.groupby('batsman')['batsman_runs'].count().reset_index()
sixes = sixes.rename(columns={'batsman':'batsman','batsman_runs':'6s'})
sixes.head()
```

```python
data = [balls['batsman'],balls['ball'],runs['batsman_runs'],fours['4s'],sixes['6s']]
headers = ['batsman','balls','batsman_runs','4s','6s']
players = pd.concat(data, axis=1, keys = headers )
players.head()
```

```python
players.dtypes
```

```python
players.info()
```

```python
players['4s'] = players['4s'].fillna(0)
players['6s'] = players['6s'].fillna(0)
```

```python
players['4s'] = players['4s'].astype('int')
players['6s'] = players['6s'].astype('int')
players.head()
```

```python
players['strike_rate'] = (players['batsman_runs']/players['balls'])*100
players.head()
```

```python
grp = df.groupby(['match_id','batsman','batting_team'])['batsman_runs'].sum().reset_index()
grp.head()
```

```python
max_score = grp.groupby('batsman')['batsman_runs'].max().reset_index()
max_score.columns = ['batsmen','best_score']
max_score.head()
```

```python
players = pd.concat([players,max_score.iloc[:,1]],axis = 1)
players.head()
```

```python
best_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False)[:10]
best_batsmen.plot(kind = 'bar')
```

```python
best_score_players = players.sort_values(by='best_score', ascending=False)
best_score_players[['batsman','best_score']].head(10)
```
